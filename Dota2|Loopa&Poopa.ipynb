```python
botan = pd.read_csv('csv/algae.csv')
```

```python
botan
```

```python
botan.groupby('genus').mean()
```

```python
mean_concentrations = botan.groupby(['genus'])\
    .aggregate( \
    mean_suc = ('sucrose', 'mean'), \
    mean_alanin = ('alanin', 'mean'), \
    mean_citrate = ('citrate', 'mean'),
    mean_glucose = ('glucose', 'mean'),
    mean_oleic_acid = ('oleic_acid', 'mean')
)  
```

```python
mean_concentrations
```

```python
botan.groupby('genus').aggregate({'alanin' : 'describe'})
```
