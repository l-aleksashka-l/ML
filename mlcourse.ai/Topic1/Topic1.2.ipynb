```python
def toFixed(numObj, digits=0):
    return f"{numObj:.{digits}f}"
```

```python
for i, sub_df in data.groupby('looks'):
    print(type(i))
    print()
    print(sub_df.head())
    print()
    print("Counts : ", sub_df.exper.count())
    print()
    print("Median : ", sub_df["wage"].median())
    print()
    print("% by female : ", toFixed(sub_df["female"].mean()*100,2), " %")
    print()
    
```
