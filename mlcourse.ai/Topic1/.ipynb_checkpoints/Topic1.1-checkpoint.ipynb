# Applying Functions to Cells, Columns and Rows

```python
df.apply(np.max)
```

```python
x = df[df["State"].apply(lambda letter : letter[0] == "W")]
```

```python
d={"Yes":True, "No":False}
df["International plan"] = df["International plan"].map(d)
df.head()
```

```python
df["International plan"].dtype
```

```python
df = df.replace({"Voice mail plan" : d})
df.head()
```

# Grouping

```python
columns_to_show = ["Total day minutes", "Total eve minutes", "Total night minutes"]
df.groupby(["Churn"])[columns_to_show].describe(percentiles=[])
```

```python
df.groupby(["Churn"])[columns_to_show].agg([np.min, np.max, np.mean, np.std])
```

# Summary tables

```python
pd.crosstab(df["Churn"], df["International plan"])
```

```python
pd.crosstab(df["Voice mail plan"], df["International plan"])
```

```python
pd.crosstab(x.State, df["International plan"])
```

```python
pd.crosstab(df["Churn"], df["Voice mail plan"], normalize=True)
```
