```python
import pandas as pd 
import numpy as np
```

```python
stud = pd.read_csv('csv/StudentsPerformance.csv')
```

```python
stud
```

```python
group = stud.groupby(['gender', 'race/ethnicity']) \
.aggregate({'math score' : 'mean', 'reading score' : 'var', 'writing score' : 'std'}) \
.rename(columns = {'math score' : 'mean_math_score', 'reading score' : 'var_reading_score', 'writing score' : \
                   'std_writing_score'})
```

```python
stud.groupby(['gender', 'race/ethnicity'], as_index=False) \
.aggregate({'math score' : 'mean', 'reading score' : 'var', 'writing score' : 'std'}) \
.rename(columns = {'math score' : 'mean_math_score', 'reading score' : 'var_reading_score', 'writing score' : \
                   'std_writing_score'})
```
